```python
import cv2
import numpy as np
import datetime  # 날짜와 시간을 기록하기 위해

# 배구 코트를 나타내는 빈 이미지 생성
court = np.zeros((450, 900, 3), np.uint8)

# 배구 코트의 주요 포인트 정의
court_lines = [
    [(0, 0), (900, 0)],     # 상단 경계선
    [(0, 450), (900, 450)], # 하단 경계선
    [(0, 0), (0, 450)],     # 왼쪽 경계선
    [(900, 0), (900, 450)], # 오른쪽 경계선
    [(450, 0), (450, 450)], # 중앙선 (네트 위치)
    [(300, 0), (300, 450)], # 왼쪽 어택라인 (네트에서 3미터)
    [(600, 0), (600, 450)]  # 오른쪽 어택라인 (네트에서 3미터)
]

# 마우스 클릭 이벤트 처리 및 특별 이벤트 저장을 위한 리스트
coord_list = []
special_events = []

def mouse_callback(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN or event == cv2.EVENT_RBUTTONDOWN:
        index = len(coord_list) + 1  # 순서는 1부터 시작
        color = (255, 0, 0) if event == cv2.EVENT_LBUTTONDOWN else (0, 0, 255)
        coord_list.append((x, y, index, color))
        print(f"기록된 {index}: ({x}, {y})")  # 콘솔에 좌표 및 순서 출력

cv2.namedWindow("court")
cv2.setMouseCallback("court", mouse_callback)

while True:
    court.fill(0)  # 코트 이미지 리셋
    for line_start, line_end in court_lines:
        cv2.line(court, line_start, line_end, (0, 255, 0), 2)
    
    # 이전에 표시된 위치 사이에 선 그리기
    if len(coord_list) > 1:
        for i in range(len(coord_list) - 1):
            cv2.line(court, (coord_list[i][0], coord_list[i][1]), 
                     (coord_list[i+1][0], coord_list[i+1][1]), (255, 255, 255), 2)

    # 마우스 클릭 위치에 원과 순서 번호 표시
    for x, y, index, color in coord_list:
        cv2.circle(court, (x, y), 10, color, -1)
        cv2.putText(court, str(index), (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 0), 1)
    
    cv2.imshow("court", court)

    key = cv2.waitKey(30) & 0xFF
    if key == 27:  # ESC 키를 누르면 종료
        break
    elif key == ord('s'):  # 's' 키를 누르면 좌표를 파일에 저장
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
        file_name = f"coordinates_{timestamp}.txt"
        with open(file_name, "w") as file:
            for x, y, index, _ in coord_list:
                file.write(f"{index}: {x},{y}\n")
        print(f"{file_name} 파일에 좌표 저장됨")
    elif key == ord('r'):  # 'r' 키를 누르면 좌표와 시퀀스 리셋
        coord_list = []
        special_events = []
        print("좌표 및 인덱스가 리셋되었습니다")
    elif key == ord('g'):  # 'g' 키를 누르면 특별 이벤트 기록
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        special_events.append(f"Setter set at {timestamp}")
        print(f"{timestamp}에 세터 설정 이벤트 기록됨")
    elif key == ord('p'):  # 'p' 키를 누르면 현재 코트 뷰를 이미지 파일로 저장
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
        cv2.imwrite(f"court_snapshot_{timestamp}.png", court)
        print(f"'court_snapshot_{timestamp}.png'로 코트 스냅샷 저장됨")
    elif key == ord('e'):  # 'e' 키를 누르면 마지막 변경 취소
        if coord_list:
            removed = coord_list.pop()
            print(f"마지막 기록된 점 제거됨: {removed}")
    elif key == ord('q'):  # 'q' 키를 누르면 실수 기록
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        special_events.append(f"Fault committed at {timestamp}")
        print(f"{timestamp}에 실수 발생 이벤트 기록됨")

cv2.destroyAllWindows()

```

```python
import re

def extract_numbers(input_string):
    # 괄호 안의 숫자 쌍을 찾는 정규 표현식 패턴입니다
    pattern = r'\(\s*(\d+)\s*,\s*(\d+)\s*\)'
    # 문자열에서 패턴에 해당하는 모든 경우를 찾습니다
    matches = re.findall(pattern, input_string)
    # 찾은 숫자 쌍을 쉼표로 구분하여 괄호 안에 다시 넣습니다
    result = ', '.join(f'({match[0]}, {match[1]})' for match in matches)
    return result

# 파일을 읽어서 처리
file_path = r"C:\Users\dka57\AppData\Local\Programs\Microsoft VS Code\coordinates_2024-05-12_17-47-29.txt"
with open(file_path, 'r', encoding='utf-8') as file:
    content = file.read()
    print("파일 내용 일부:", content[:500])  # 파일 내용의 처음 500자 출력

# 결과 출력
result = extract_numbers(content)
print("추출된 결과:", result)

# 결과를 새 파일에 저장
with open('output.txt', 'w', encoding='utf-8') as output_file:
    output_file.write(result)
```

```python
import cv2
import numpy as np

# 파일에서 좌표 데이터 읽기
file_path = "C:\\Users\\dka57\\AppData\\Local\\Programs\\Microsoft VS Code\\output.txt"
with open(file_path, 'r') as file:
    content = file.read()
    coords = eval(content)  # 파일의 내용을 파이썬 리스트로 변환

# 배구 코트를 나타내는 빈 이미지 생성
court = np.zeros((450, 900, 3), dtype=np.uint8)

# 배구 코트의 주요 포인트 정의
court_lines = [
    [(0, 0), (900, 0)],     # 상단 경계선
    [(0, 450), (900, 450)], # 하단 경계선
    [(0, 0), (0, 450)],     # 왼쪽 경계선
    [(900, 0), (900, 450)], # 오른쪽 경계선
    [(450, 0), (450, 450)], # 중앙선 (네트 위치)
    [(300, 0), (300, 450)], # 왼쪽 어택라인 (네트에서 3미터)
    [(600, 0), (600, 450)]  # 오른쪽 어택라인 (네트에서 3미터)
]

# 코트 선 그리기
for line_start, line_end in court_lines:
    cv2.line(court, line_start, line_end, (255, 255, 255), 2)  # 흰색 선

# 좌표에 따른 활동 위치 그리기
for x, y in coords:
    cv2.circle(court, (x, y), 5, (0, 255, 0), -1)  # 활동 위치를 녹색 점으로 표시

# 결과 이미지 표시
cv2.imshow('Volleyball Court Activity', court)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
